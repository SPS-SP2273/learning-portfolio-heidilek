## Task 4

```python
def rename_my_file(old_file_name):
    month = {
    'jan': '01', 'feb': '02', 'mar': '03',
    'apr': '04', 'may': '05', 'jun': '06',
    'jul': '07', 'aug': '08', 'sep': '09',
    'oct': '10', 'nov': '11', 'dec': '12'
    }
    mm, remainder = old_file_name.split("-")
    new_file_name = month[mm] + "-" + remainder
    if old_file_name != new_file_name:
        return new_file_name
    else:
        return old_file_name
    
print(rename_my_file("oct-08_singapor.txt"))
```

```python
files = glob.glob('os-collaboration-exercise-data*/*.txt')
file_names_only_list = []
for file in files:
    file_name = file.split("\\")[1]
    file_names_only_list.append(file_name)
    
for file in file_names_only_list:
    new_file_name = rename_my_file(file)
    old_path = os.path.join('os-collaboration-exercise-data', file)
    new_path = os.path.join('os-collaboration-exercise-data', new_file_name)
    os.rename(old_path, new_path)
    print(f"{file} ----> {new_file_name}")
```

## Task 5

```python
cities = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

for city in cities:
    try:
        os.mkdir(city)
        print(f'Creating {city}')
    except FileExistsError:
        print(f'{city} already exists; skipping creation.')
```

## Task 6

```python
cairo_files = glob.glob('os-collaboration-exercise-data*/*cairo*')
for file in cairo_files:
    shutil.copy(file, 'Cairo')
    print('Copied file to Cairo')
```

## Task 7

```python
other_cities = ['Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

for city in other_cities:
    city_files = glob.glob(f'os-collaboration-exercise-data*/*{city.lower()}*')
    for file in city_files:
        shutil.copy(file, city)
        print(f'Copied file to {city}')
```
